## 1. Mean Square Error

Find the mean square error between target array and predict array.<br>
You can find how to calculate mean square error [here.](https://en.wikipedia.org/wiki/Mean_squared_error)

Example
```python
target = np.array([1, 2, 1, 2, 3, 1, 2])
predict = np.array([2, 1, 2, 2, 1, 2, 1])
```
`mean_square_error(target, predict)` --> `1.2857142857142858`

````python
# 1 mean square error
target = np.array([1, 2, 1, 2, 3, 1, 2])
predict = np.array([2, 1, 2, 2, 1, 2, 1])

def mean_square_error(target, predict):
    import numpy as np
    '''
    Calculate Mean Square Error Between two given arrays.
    Inputs:
        target(ndarray) --> Target array
        predict(ndarray) --> Predict array
    Return:
        mse(float64) --> mean square error value
    '''
    ####CODE HERE####
    
    n = len(target)
    n1= len(predict)
    
    if n1 == n:
        mse = (np.square(target-predict)).mean()
  
    else:
        mse = 'Target list and Predict list do not have the same number of elements'
        
    #################
    return mse
````

````python
mse = mean_square_error(target,predict)
print(type(mse))
print(mse)
````

## 2. Stack three channels Images

Given inputs are three separate color channels, red, green and blue channel. Return the corresponding Color Image. 

Output Example

```python
rgb_image = stack_channels(r, g, b)
print(rgb_image.shape)
plt.imshow(rgb_image)
plt.show()
```
```
(900, 900, 3)
```
![](images/cvml.png)

````python
# 2. Stack three channels images
def stack_channels(red, green, blue):
    import numpy as np
    '''
    Stack three channels images and return color image.
    Inputs:
        red(ndarray) --> red.shape = (500, 1620)
        green(ndarray) --> green.shape = (900, 900)
        blue(ndarray) --> blue.shape = (900, 900)
    Return:
        rgb_image(ndarray) --> rgb_image.shape = (900,900,3)
    '''
    ####CODE HERE####
    
    red = red.reshape((900,900))
    rgb_image = np.dstack((red,green,blue))
    
    #################
    return rgb_image
````

````python
with open("cvml.pkl", 'rb') as reader:
    arrays = pickle.load(reader)
    
r = arrays[0]
g = arrays[1]
b = arrays[2]

rgb_image = stack_channels(r, g, b)
print(rgb_image.shape)
plt.imshow(rgb_image)
plt.show()
````

## 3. Plot Sigmoid Function

Sigmoid function is often used in Deep Learning as the activation function. In this exercise, you will have to study the equation of the sigmoid function and plot it with matplotlib function.

Output example

````python
# plot sigmoid function
x = np.linspace(-10, 10, 100)

def sigmoid(x):
    import numpy as np
    import matplotlib.pyplot as plt
    '''
    find output of sigmoid function and plot it.
    Inputs:
        x(ndarray : float64) --> input data points
    Return:
        y(ndarray : float64) --> Sigmoid funtion output : same shape as input data points
    '''
    ####CODE HERE####
    
    y = 1/(1+np.exp(-x))
    plt.plot(x,y)
    
    #################
    return y
````

## 4. Image Manipulation with OpenCV

Complete the following small exercises to complete one whole process. main function here will be provided for overall process. And when calling the main function here as the example, the output should be:

```python
def main():
```

![](images/rotated.jpg)

````python
# playing with cv2

def read_img(file_name):
    '''
    Read and return 'RGB' image array.
    Input:
        file_name(str) --> image name.
    Return:
        img_array(ndarray) --> RGB image array.
    '''
    ####CODE HERE####
    
    img_array = mpimg.imread(file_name)
    
    #################
    return img_array


def get_green_channel(img):
    '''
    Slicing Color Image to get Green channel Image.
    Input:
        img(ndarray) --> RGB image array.
    Return:
        green(ndarray) --> shape = (h, w)
    '''
    ####CODE HERE####
    
    green = img[:,:,1]
    
    #################
    return green


def resize_img(img, h, w):
    '''
    Resize the image by the given shape.
    Inputs:
        img(ndarray) --> 2D input image
        h(Int), w(Int) --> resized shape
    Return
        resized(ndarray) --> resized image, shape=(h, w)
    '''
    ####CODE HERE####
    
    resized = cv2.resize(img, (w,h))
    
    #################
    return resized


def rotate_img(img, degree, scale=1.0):
    '''
    Rotate the given image by certain degree.
    Inputs:
        img(ndarray) --> 2D image to be rotated
        degree(Int) --> rotational angle, in degree.
        scale(float) --> Scale of the rotated image, default is one.
    Return:
        rotated(ndarray) --> 2D Rotated image of the input image.
    '''
    ####CODE HERE####
    
    h,w = img.shape
    M = cv2.getRotationMatrix2D((w//2, h//2), degree, scale)
    rotated = cv2.warpAffine(img, M, (w, h))
    
    #################
    return rotated
````

````python
def main():
    img = read_img('images/cvml.png')
    green = get_green_channel(img)
    resized = resize_img(green, 600, 600)
    rotated = rotate_img(resized, 60, 0.5)
    plt.imshow(rotated, cmap='gray')
    plt.show()

main()
````

## 5. Stack Two Images

The logo is divided in half, and you need to restore it to be the complete logo. The output is the Logo the same as in [question 2](#rgb).

````python
# Stack left and right logo.
with open('half_cvml.pkl', 'rb') as reader:
    arrays = pickle.load(reader)
    
left_logo = arrays[0]
right_logo = arrays[1]


def stack_images(half1, half2):
    import numpy as np
    '''
    Stack two images to complete the logo.
    Inputs:
        half1(ndarray) --> left side of logo
        half2(ndarray) --> right side of logo
    Return
        logo(ndarray) -> complete logo
    
    '''
    ####CODE HERE####
    
    logo = np.hstack((half1, half2))
    
    #################
    return logo
````

````python
# Output example
 
plt.figure(figsize=(10,10))
plt.imshow(stack_images(left_logo, right_logo))
plt.show()
````
